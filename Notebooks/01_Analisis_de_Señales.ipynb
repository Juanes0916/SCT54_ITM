### Ejemplo 1 <br>
a) Determinar las amplitudes máximas y las frecuencias de las primeras cinco armónicas impares. b) Trazar el espectro de frecuencias. c) Calcular el voltaje instantáneo total, para varios tiempos, y trazar la forma de onda en el dominio del tiempo.

![texto alternativo](https://github.com/cdguarnizo/SCT54_ITM/blob/master/Notebooks/figs/Tomasi_eje_1-12.png?raw=true)

```
#Datos
T = 1e-3
V = 4.

#Caculos
f = 1/T   #Frecuencia fundamental
n = np.arange(1,10,2) #Vector de armonicos
print(n)
```

```
#Solucion parte a)
fn = f*n
Vn = 4.0*V/(np.pi*n)
print("fn=",fn,'\n',"Vn=",Vn)
```

```
#Solucion parte b)
plt.stem(fn,Vn)
```

```
#Solucion parte c)
t = np.linspace(0,T,100)
Vt = 0.
for fi in fn:
  Vt += V*np.sin(2.0*np.pi*fi*t)

plt.plot(t,Vt)
```

### Ejemplo 2
a) Determinar la componente de corriente directa (DC). b) Determine las amplitudes maximas de las 10 primeras armónicas. c) Grafique la funcion $\sin(x)/x$, con $x=\pi(\tau/T)$. Trace el espectro de frecuencias.

![texto alternativo](https://github.com/cdguarnizo/SCT54_ITM/blob/master/Notebooks/figs/Tomasi_eje_1-15.png?raw=true)

```
#Datos
tau = .4e-3
T = 2e-3
V = 1.0

DC = tau/T
```

```
# Solucion parte a) Componente DC
V0 = V*DC
print(V0)
```

```
# Solucion parte b) 
n = np.arange(1,10+1)
x = np.pi*DC
f = 1/T
fn = f*n
Vn = 2*V0*np.sin(n*x)/(n*x)

print(fn,Vn)
```

```
# Solucion parte c) ]
plt.plot(fn,Vn)
```

```
# Solucion parte d)
fn = np.hstack((0,fn))
Vn = np.hstack((V0,Vn))
plt.stem(fn,np.abs(Vn))
```
